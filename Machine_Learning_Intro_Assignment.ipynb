# **Question 1:** Explain the differences between AI, ML, Deep Learning (DL), and Data Science (DS).

Artificial Intelligence (AI), Machine Learning (ML), Deep Learning (DL), and Data Science (DS) are interrelated fields, but each has distinct objectives and scope.

**Artificial Intelligence (AI):**

AI is the broadest concept that focuses on creating machines capable of mimicking human intelligence. It involves enabling systems to perform tasks like reasoning, learning, problem-solving, decision-making, and natural language understanding.
* For example, virtual assistants like Siri and Alexa are AI systems that can understand and respond to human queries.

**Machine Learning (ML):**

ML is a subset of AI that allows systems to learn from data and improve performance over time without being explicitly programmed. Instead of hard-coding rules, ML algorithms identify patterns and make predictions.
* For instance, spam email detection is an ML application where the system learns from past labeled emails to classify new ones.

**Deep Learning (DL):**

DL is a specialized subfield of ML that uses artificial neural networks with many layers (deep architectures) to model complex patterns in data. It is particularly effective for tasks like image recognition, speech processing, and natural language translation.
* For example, facial recognition systems used in smartphones rely on deep learning.

**Data Science (DS):**

Data Science is an interdisciplinary field that uses statistics, computer science, domain knowledge, and ML to extract insights and build predictive or prescriptive models from structured and unstructured data. It includes the full pipeline of data collection, cleaning, analysis, visualization, and modeling.
* For example, a data scientist in retail may analyze customer purchase data to recommend products.

**In summary:**

* AI is the overarching goal of building intelligent systems.

* ML is one way to achieve AI by enabling systems to learn from data.

* DL is an advanced subset of ML using deep neural networks.

DS is the broader discipline of analyzing and interpreting data, often applying ML/DL techniques for business and scientific insights.

# **Question 2:** What are the types of machine learning? Describe each with one real-world example.

Machine learning can be broadly classified into three main types: Supervised Learning, Unsupervised Learning, and Reinforcement Learning. Each has different approaches, techniques, and applications.

**1. Supervised Learning:**

In supervised learning, the model is trained using labeled data, meaning that both input features and their corresponding correct outputs (targets) are known. The goal is to learn a mapping function that can predict outputs for unseen data.

*Real-world Example:* Predicting house prices based on features such as location, size, and number of rooms. The dataset includes historical data with actual house prices (labels), which helps the model learn.

**2. Unsupervised Learning:**

Unsupervised learning deals with unlabeled data, where the system tries to find hidden patterns, groupings, or structures without predefined outputs. The model organizes the data or reduces its dimensionality.

*Real-world Example:* Customer segmentation in marketing. Retail companies use clustering algorithms like K-Means to group customers based on purchasing behavior, allowing targeted promotions.

**3. Reinforcement Learning (RL):**

Reinforcement learning is based on agents interacting with an environment. The agent learns through trial and error by receiving rewards (positive feedback) or penalties (negative feedback). The goal is to maximize cumulative rewards over time.

*Real-world Example:* Self-driving cars use reinforcement learning to make decisions such as when to accelerate, stop, or change lanes by receiving feedback from sensors and the environment.

**Additional Types (sometimes considered subcategories):**

**Semi-Supervised Learning:** Uses a mix of labeled and unlabeled data. Example: Google Photos using a few tagged images to organize a large photo library.

**Online Learning:** Model updates continuously as new data arrives, useful for stock market prediction or real-time fraud detection.

**In summary:**

* Supervised Learning → Works with labeled data to predict outcomes.

* Unsupervised Learning → Works with unlabeled data to discover hidden structures.

* Reinforcement Learning → Works with an environment to maximize rewards through feedback.

These types together cover the majority of real-world applications of machine learning.

# **Question 3:** Define overfitting, underfitting, and the bias-variance tradeoff in machine learning.

In machine learning, model performance depends on how well it generalizes to unseen data. Three key concepts that describe this are overfitting, underfitting, and the bias-variance tradeoff.

**Overfitting:**

Overfitting occurs when a model learns the training data too well, including noise, random fluctuations, and irrelevant details. As a result, the model performs very well on training data but poorly on new or test data. This happens when the model is too complex, such as using too many parameters or deep trees in decision trees.

*Example:* A model predicting exam scores memorizes the exact scores of students instead of learning general patterns like study hours and performance.

**Underfitting:**

Underfitting happens when the model is too simple to capture the underlying structure of the data. It fails to learn important patterns, resulting in poor accuracy on both training and testing data. This usually occurs when using a weak model, insufficient training, or ignoring important features.

*Example:* Using a straight-line regression to model a highly curved relationship between variables results in underfitting.

**Bias-Variance Tradeoff:**

Bias and variance are two sources of error in machine learning:

* Bias is the error introduced when the model is too simple and makes strong assumptions about the data. High bias leads to underfitting.

* Variance is the error introduced when the model is too sensitive to small changes in the training data. High variance leads to overfitting.

The bias-variance tradeoff refers to finding the optimal balance between the two. A good model should neither underfit nor overfit; instead, it should achieve low bias and low variance, ensuring accurate predictions on both training and unseen datasets.

**In summary:**

* Overfitting → Model too complex → Good on training data, poor on new data.

* Underfitting → Model too simple → Poor on both training and new data.

* Bias-Variance Tradeoff → Balancing simplicity and complexity to achieve generalization.

# **Question 4:** What are outliers in a dataset, and list three common techniques for handling them.

Outliers are data points that significantly deviate from the overall pattern of a dataset. They can be unusually high or low values compared to the majority of observations. Outliers may occur due to data entry errors, measurement mistakes, sampling variability, or they may represent rare but valid events.

**Why outliers matter:**

* They can distort statistical measures like mean and standard deviation.

* They may reduce the accuracy of machine learning models by skewing the training process.

* Sometimes, outliers may carry meaningful insights (e.g., fraud detection in banking).

**Three common techniques for handling outliers are:**

**Removal of Outliers:**

If outliers are due to errors or do not contribute meaningful information, they can be removed. This is commonly done when the dataset is large, and removing a few points does not affect representativeness.

*Example:* Deleting impossible age values like 150 years in a customer dataset.

**Transformation:**

Applying mathematical transformations reduces the influence of extreme values. Logarithmic, square root, or Box-Cox transformations compress large deviations, making the data more normally distributed.

*Example:* Applying log transformation on income data where some individuals earn extremely high salaries.

**Capping (Winsorization):**

Extreme values are replaced with certain percentile thresholds (e.g., replacing all values above the 95th percentile with the 95th percentile value). This reduces the impact of outliers while keeping them within a realistic range.

*Example:* Setting all extreme transaction amounts above the 99th percentile to the 99th percentile value in a retail dataset.

**In summary:**

Outliers are extreme data points that can affect analysis and model performance. The most common ways to handle them are removal, transformation, and capping, though the choice depends on whether the outliers are errors or valuable rare cases.

# **Question 5:** Explain the process of handling missing values and mention one imputation technique for numerical and one for categorical data.

Missing values occur in datasets when no value is recorded for certain attributes. They are common in real-world data due to reasons like incomplete surveys, equipment malfunction, or human error. If not handled properly, missing values can reduce the quality of analysis and lead to biased or inaccurate machine learning models.

**Process of Handling Missing Values:**

Identification of Missing Data:

* The first step is to detect missing values using methods like isnull() in Python or descriptive statistics.

**Understanding the Pattern:**

Determine if the missingness is:

* MCAR (Missing Completely at Random): Missing values are unrelated to any data.

* MAR (Missing at Random): Missingness depends on observed data.

* MNAR (Missing Not at Random): Missingness depends on unobserved data.

**Choosing a Strategy:**

**Deletion Methods:** Remove rows or columns with excessive missing values (useful when the dataset is large and missingness is small).

**Imputation Methods:** Replace missing values with estimated values to preserve data size.

**Imputation Techniques:**

**For Numerical Data:** One common method is Mean Imputation, where missing values are replaced by the mean of the non-missing values.
Example: If income values are missing, replace them with the average income of the dataset.

**For Categorical Data:** A widely used technique is Mode Imputation, where missing values are replaced with the most frequent category in the dataset.

*Example:* If the “Payment Method” column has missing entries, fill them with the most common method (e.g., “Credit Card”).

**In summary:**

Handling missing values involves identifying missing data, understanding why it occurs, and applying appropriate techniques. For numerical variables, mean imputation is commonly used, while for categorical variables, mode imputation is an effective solution. These techniques ensure the dataset remains consistent and reliable for machine learning models.

# **Question 8:** Write a Python script to:

● Generate 1000 samples from a normal distribution.

● Introduce 50 random missing values.

● Fill missing values with the column mean.

● Plot a histogram before and after imputation.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Generate 1000 samples from a normal distribution
data = np.random.normal(loc=50, scale=10, size=1000)  # mean=50, std=10
df = pd.DataFrame(data, columns=['Value'])

# Step 2: Introduce 50 random missing values
nan_indices = np.random.choice(df.index, 50, replace=False)
df.loc[nan_indices, 'Value'] = np.nan

# Step 3: Plot histogram before imputation
plt.hist(df['Value'].dropna(), bins=30, color='skyblue', edgecolor='black')
plt.title("Histogram Before Imputation")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()

# Step 4: Fill missing values with the column mean
df['Value'].fillna(df['Value'].mean(), inplace=True)

# Step 5: Plot histogram after imputation
plt.hist(df['Value'], bins=30, color='lightgreen', edgecolor='black')
plt.title("Histogram After Imputation")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

# **Question 10:** You are working as a data scientist for a retail company. You receive a customer transaction dataset that contains:

● Missing ages,

● Outliers in transaction amount,

● A highly imbalanced target (fraud vs. non-fraud),

● Categorical variables like payment method.

Explain the step-by-step data preparation plan you’d follow before training a machine learning model. Include how you’d address missing data, outliers, imbalance, and encoding.

**Step 1:** Handling missing ages – Use imputation (mean or median for numerical values). Alternatively, predictive imputation (e.g., regression) could be applied if features like location and spending habits are correlated with age.

**Step 2:** Handling outliers in transaction amounts – Detect using IQR (interquartile range) or z-score method. Outliers can be capped at certain percentiles or transformed using log scaling.

**Step 3:** Handling imbalanced target – Apply techniques like oversampling the minority class (SMOTE), undersampling the majority class, or using class-weighted algorithms.

**Step 4:** Handling categorical variables – Use encoding methods. For nominal features like payment method, apply one-hot encoding. For ordinal categories, label encoding is appropriate.

**Step 5:** Final preprocessing – Apply scaling/normalization to numerical features, split dataset into training and testing sets, and ensure data pipeline consistency.

```python
import numpy as np
import pandas as pd

# Create a synthetic dataset mimicking the retail transaction data
np.random.seed(42)  # for reproducibility
n_samples = 1000

# Generate numerical features
age = np.random.randint(18, 70, size=n_samples).astype(float) # Create as float to allow NaN
amount = np.random.uniform(10, 500, size=n_samples)

# Introduce some missing values in age
missing_age_indices = np.random.choice(n_samples, size=int(0.1 * n_samples), replace=False)
age[missing_age_indices] = np.nan

# Introduce some outliers in amount
outlier_indices = np.random.choice(n_samples, size=int(0.02 * n_samples), replace=False)
amount[outlier_indices] = amount[outlier_indices] * np.random.uniform(2, 10, size=len(outlier_indices))

# Generate categorical feature
payment_methods = ['Credit Card', 'Debit Card', 'Cash', 'Online Transfer']
payment_method = np.random.choice(payment_methods, size=n_samples, p=[0.6, 0.2, 0.1, 0.1])

# Generate imbalanced target variable (fraud vs. non-fraud)
is_fraud = np.random.choice([0, 1], size=n_samples, p=[0.95, 0.05]) # 5% fraud

# Create DataFrame
df = pd.DataFrame({
    'age': age,
    'amount': amount,
    'payment_method': payment_method,
    'is_fraud': is_fraud
})

print("Synthetic DataFrame created:")
display(df.head())
print("\nClass distribution of 'is_fraud':")
display(df['is_fraud'].value_counts())
```

```python
# ====== Imports

from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, RobustScaler, FunctionTransformer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score, average_precision_score

# If you want resampling:
# from imblearn.pipeline import Pipeline as ImbPipeline
# from imblearn.over_sampling import SMOTE

# ====== Example: assume df with columns
#   numeric: ["age", "amount", ...]
#   categorical: ["payment_method", ...]
#   target: "is_fraud"
# Replace with your real column lists.
numeric_features = ["age", "amount"]
categorical_features = ["payment_method"]

X = df[numeric_features + categorical_features]
y = df["is_fraud"].astype(int)

# ====== Train/test split (stratified)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

# ====== Numeric pipeline: median impute, optional log on 'amount', robust scale
def log_amount_only(X_df):
    X_copy = X_df.copy()
    if "amount" in X_copy:
        X_copy["amount"] = np.log1p(X_copy["amount"].astype(float))
    return X_copy

numeric_pipeline = Pipeline(steps=[
    ("impute", SimpleImputer(strategy="median")),
    ("log_amount", FunctionTransformer(log_amount_only, feature_names_out="one-to-one")),
    ("scale", RobustScaler())
])

# ====== Categorical pipeline: mode impute + OHE (ignore unknowns at inference)
categorical_pipeline = Pipeline(steps=[
    ("impute", SimpleImputer(strategy="most_frequent")),
    ("ohe", OneHotEncoder(handle_unknown="ignore"))
])

# ====== ColumnTransformer
preprocess = ColumnTransformer(
    transformers=[
        ("num", numeric_pipeline, numeric_features),
        ("cat", categorical_pipeline, categorical_features),
    ]
)

# ====== Class-weighted baseline model
clf = LogisticRegression(
    max_iter=2000,
    class_weight="balanced",
    n_jobs=None  # set if using sklearn that supports it
)

# ====== Final pipeline without resampling
model = Pipeline(steps=[
    ("prep", preprocess),
    ("clf", clf)
])

# ====== Fit & basic evaluation
model.fit(X_train, y_train)
y_pred_proba = model.predict_proba(X_test)[:, 1]
print("ROC-AUC:", roc_auc_score(y_test, y_pred_proba))
print("PR-AUC :", average_precision_score(y_test, y_pred_proba))

# ====== Optional: use SMOTE after preprocessing (commented)
# from imblearn.pipeline import Pipeline as ImbPipeline
# from imblearn.over_sampling import SMOTE
# smote_model = ImbPipeline(steps=[
#     ("prep", preprocess),
#     ("smote", SMOTE(random_state=42)),
#     ("clf", clf)
# ])
# smote_model.fit(X_train, y_train)
# ypp = smote_model.predict_proba(X_test)[:, 1]
# print("ROC-AUC (SMOTE):", roc_auc_score(y_test, ypp))
# print("PR-AUC  (SMOTE):", average_precision_score(y_test, ypp))
```
